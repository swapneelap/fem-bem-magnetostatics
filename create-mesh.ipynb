# Create mesh

```python
import gmsh
```

```python
L = 50
l_c = 2
```

```python
gmsh.initialize()
model = gmsh.model()
model.add("Cube")
```

```python
cube = model.occ.add_box(-L/2, -L/2, -L/2, L/2, L/2, L/2)
```

```python
cube
```

```python
model.occ.synchronize()
```

```python
boundaries = model.get_boundary([(3, cube)], oriented=False)
```

```python
boundaries
```

```python
model.add_physical_group(3, [cube], name="Material");
```

```python
model.add_physical_group(2, [e[1] for e in boundaries], name="Boundary");
```

```python
model.mesh.set_size(model.get_entities(0), l_c)
```

```python
model.mesh.generate(3)
```

```python
# gmsh.fltk.run()
```

```python
gmsh.write("cube.msh")
gmsh.finalize()
```
